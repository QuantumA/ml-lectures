```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_classification
from classifiers import Perceptron
from classifiers import Adaline
import matplotlib.gridspec as gridspec

from visual.tools import plot_decision_regions
```

```python
gs = gridspec.GridSpec(2, 2)
X, y = make_moons(n_samples=10000, random_state=123)
fig = plt.figure(figsize=(10,8))
```

#### Perceptron

```python
ppn = Perceptron(eta=0.05, random_state=1)
ppn.fit(X, y, epochs=50)


fig = plot_decision_regions(X=X, y=y, classifier=ppn)
plt.title("Perceptron")
```

```python
X, y = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)

ppn = Perceptron(eta=0.05, random_state=1)
ppn.fit(X, y, epochs=50)
fig = plt.figure(figsize=(10,8))


fig = plot_decision_regions(X=X, y=y, classifier=ppn)
plt.title("Perceptron")
```

#### Adaline


```python
alne = Adaline(lr=0.0001, random_state=1)
alne.fit(X, y, epochs=50)


fig = plot_decision_regions(X=X, y=y, classifier=alne)
plt.title("Adaline")
```

```python
plt.plot(alne.losses, color='green', marker="s", markersize=1, label="Train loss")
plt.title("Loss curves")
plt.legend(loc="upper left")
plt.ylabel('Loss')
plt.xlabel('No. epoch')
plt.show()
```
