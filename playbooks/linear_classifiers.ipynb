```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons, make_classification
from classifiers import Perceptron
from classifiers.adaline import AdalineGD
from classifiers.adaline_sgd import AdalineSGD
import matplotlib.gridspec as gridspec

from visual.tools import plot_decision_regions
```

```python
gs = gridspec.GridSpec(2, 2)
X, y = make_moons(n_samples=100, random_state=123)
fig = plt.figure(figsize=(10,8))
```

#### Perceptron

```python
ppn = Perceptron(eta=0.05, random_state=1)
ppn.fit(X, y, epochs=50)


fig = plot_decision_regions(X=X, y=y, classifier=ppn)
plt.title("Perceptron")
```

```python
X, y = make_classification(n_features=2,
                           n_redundant=0,
                           n_informative=1,
                           n_clusters_per_class=1,
                           random_state=1
                           )

ppn = Perceptron(eta=0.01, random_state=1)
ppn.fit(X, y, epochs=50)
fig = plt.figure(figsize=(10,8))


fig = plot_decision_regions(X=X, y=y, classifier=ppn)
plt.title("Perceptron")
```

#### Adaline


```python
alne = AdalineGD(lr=0.01, random_state=1)
alne.fit(X, y, epochs=50)
fig = plt.figure(figsize=(10,8))

fig = plot_decision_regions(X=X, y=y, classifier=alne)
plt.title("Adaline")
```

### Adaline Vs Perceptron

##### As you can see convergence is much faster in Adaline

```python
plt.plot(alne.losses, color='green', marker="s", markersize=1, label="Train loss")
plt.title("Loss curve")
plt.legend(loc="upper right")
plt.ylabel('Loss')
plt.xlabel('No. epoch')
plt.show()
```

```python
plt.plot(ppn.errors, color='red', marker="s", markersize=1, label="Train loss")
plt.title("Updates per epoch")
plt.legend(loc="upper right")
plt.ylabel('Nº updates')
plt.xlabel('Nº epoch')
plt.show()
```

#### Adaline with Stochastic Gradient Descent

```python
alne_sgd = AdalineSGD(lr=0.005, random_state=1)
alne_sgd.fit(X, y, epochs=15)
fig = plt.figure(figsize=(10,8))
```

```python
fig = plot_decision_regions(X=X, y=y, classifier=alne_sgd)
plt.title("Adaline SGD")
plt.show()
```

```python


plt.plot(range(1, len(alne_sgd.losses) + 1), alne_sgd.losses, marker='o')
plt.xlabel("Epochs")
plt.ylabel("Average loss")
plt.tight_layout()
plt.show()

```
